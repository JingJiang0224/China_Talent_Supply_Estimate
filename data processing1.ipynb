Data Processing - Population Data

```python
import pandas as pd
```

```python
# reading the CSV file
csvFile = pd.read_csv('tmp.csv')
```

```python
csvFile.head()
```

```python
#change column name as 'col'
csvFile.columns = ['col']
```

```python
csvFile.columns
```

```python
# split 'col' into multiple columns

split_columns_adjusted = [
    "Province", "Population", "Male", "Female", "<15", "15-64", ">65", "College and Above", "Senior Secondary School", "Junior Secondary School", "Primary School"
]

df_split_adjusted = csvFile["col"].str.split(";", expand=True)
df_split_adjusted = df_split_adjusted.iloc[:, :len(split_columns_adjusted)]
df_split_adjusted.columns = split_columns_adjusted

df_split_adjusted
```

```python
df_split_adjusted.dtypes
```

```python
# Converting columns from object to float
df_split_adjusted["Province"] = df_split_adjusted["Province"].astype(str)
df_split_adjusted["Population"] = df_split_adjusted["Population"].astype(float)
df_split_adjusted["Male"] = df_split_adjusted["Male"].astype(float)
df_split_adjusted["Female"] = df_split_adjusted["Female"].astype(float)
df_split_adjusted["<15"] = df_split_adjusted["<15"].astype(float)
df_split_adjusted["15-64"] = df_split_adjusted["15-64"].astype(float)
df_split_adjusted[">65"] = df_split_adjusted[">65"].astype(float)
df_split_adjusted["College and Above"] = df_split_adjusted["College and Above"].astype(float)
df_split_adjusted["Senior Secondary School"] = df_split_adjusted["Senior Secondary School"].astype(float)
df_split_adjusted["Junior Secondary School"] = df_split_adjusted["Junior Secondary School"].astype(float)
df_split_adjusted["Primary School"] = df_split_adjusted["Primary School"].astype(float)




df_split_adjusted.dtypes
```

```python
df_split_adjusted.to_excel("new_population.xlsx")
```
