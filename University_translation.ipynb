```python
import openai
import pandas as pd
```

```python
# Load data from an Excel file
def load_data(filepath):
    return pd.read_excel(filepath, engine='openpyxl')

# Save data to an Excel file
def save_data(df, filepath):
    df.to_excel(filepath, index=False, engine='openpyxl')

# Function to translate a list of names using OpenAI's API
def translate_names(names):
    openai.api_key = 'key***'
    translations = []
    for name in names:
        try:
            response = openai.Completion.create(
                model="gpt-4",  # specify the GPT-4 model
                prompt=f"Translate the Chinese university name '{name}' into English:",
                max_tokens=60,
                n=1,
                stop=None,
                temperature=0.3
            )
            translated_name = response.choices[0].text.strip()
            translations.append(translated_name)
        except Exception as e:
            print(f"Failed to translate {name}: {str(e)}")
            translations.append(name)  # Append original name on failure
    return translations

# Main function to load, translate, and save university names
def main():
    input_filepath = 'tmp_school_names.xlsx'
    output_filepath = 'tmp_translated_school_names.xlsx'

    df = load_data(input_filepath)
    names = df['School_name'].tolist()
    translated_names = translate_names(names)
    df['Translated Names'] = translated_names
    save_data(df, output_filepath)

    print("Translation complete. Translated names saved to:", output_filepath)

# Run the main function
if __name__ == "__main__":
    main()
```
