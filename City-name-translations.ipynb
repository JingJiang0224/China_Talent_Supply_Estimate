```python
def excel_to_csv(input_excel_path, output_csv_path):
    # Load the Excel file
    df = pd.read_excel(input_excel_path)

    # Save to CSV with UTF-8 encoding
    df.to_csv(output_csv_path, encoding='utf-8', index=False)

# Example usage
excel_to_csv('tmp_school_city_name.xlsx', 'tmp_school_city_name.csv')
```

```python
import googletrans
from googletrans import Translator, LANGUAGES

# Function to translate city names using Google Translate
def translate_school(school_name, translator=None):
    if not translator:
        translator = Translator()
    translation = translator.translate(school_name, src='zh-cn', dest='en')
    return translation.text


# Create a Translator object
translator = Translator()


```

```python
# Translate each school in the DataFrame
df['English_School_Name'] = df['School_name'].apply(lambda x: translate_school(x, translator))
```
