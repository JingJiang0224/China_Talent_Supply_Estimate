```python
import pandas as pd
```

```python
def excel_to_csv(input_excel_path, output_csv_path):
    # Load the Excel file
    df = pd.read_excel(input_excel_path)
    
    # Save to CSV with UTF-8 encoding
    df.to_csv(output_csv_path, encoding='utf-8', index=False)

# Example usage
excel_to_csv('school-world-ranking.xlsx', 'school-world-ranking.csv')
```

```python
# reading the CSV file
df = pd.read_csv('school-world-ranking.csv')
```

```python
df
```

```python
# copy the data 
df_copy_data = df.copy() 
  
# apply normalization techniques 

df_copy_data['THE-WUR'] = 10 - (df_copy_data['THE-WUR'] / df_copy_data['THE-WUR'].max()) * 10 
df_copy_data['CWUR'] = 10 - (df_copy_data['CWUR'] / df_copy_data['CWUR'].max()) * 10
df_copy_data['QS-WUR'] = 10 - (df_copy_data['QS-WUR'] / df_copy_data['QS-WUR'].max()) * 10
df_copy_data['Shanghai-Ranking'] = 10 - (df_copy_data['Shanghai-Ranking'] / df_copy_data['Shanghai-Ranking'].max()) * 10
df_copy_data['Chinese Project-985'] = (df_copy_data['Chinese Project-985'] / df_copy_data['Chinese Project-985'].max()) * 10

```

```python
df_copy_data
```

```python
df_copy_data.to_csv('school-world-ranking-normalization.csv', index=False)
```
