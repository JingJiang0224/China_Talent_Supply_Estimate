```python
import pandas as pd
```

```python
def excel_to_csv(input_excel_path, output_csv_path):
    # Load the Excel file
    df = pd.read_excel(input_excel_path)

    # Save to CSV with UTF-8 encoding
    df.to_csv(output_csv_path, encoding='utf-8', index=False)

# Example usage
excel_to_csv('Count number of companies.xlsx', 'Count number of companies.csv')
```

```python
df = pd.read_csv('Count number of companies.csv')
```
