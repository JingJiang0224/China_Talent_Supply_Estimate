```python
import pandas as pd
```

```python
#### change excel into csv

def excel_to_csv(input_excel_path, output_csv_path):
    # Load the Excel file
    df = pd.read_excel(input_excel_path)

    # Save to CSV with UTF-8 encoding
    df.to_csv(output_csv_path, encoding='utf-8', index=False)

# Example usage
excel_to_csv('python _procee_doublefirstclass_disciplines.xlsx', 'python _procee_doublefirstclass_disciplines.csv')
```

```python
# reading the CSV file
df = pd.read_csv('python _procee_doublefirstclass_disciplines.csv')
```

```python
# split column "discipline"
df['discipline'] = df['discipline'].str.split('、')

# transform one row to multiple rows
transformed_df = df.explode('discipline')
```

```python
transformed_df
```

```python
###### create translator

from googletrans import Translator, LANGUAGES
import time

# Function to translate city names using Google Translate
def translate_discipline(name, translator=None):
    if not translator:
        translator = Translator()
    translation = translator.translate(name, src='zh-cn', dest='en')
    time.sleep(1)
    return translation.text


# Create a Translator object
translator = Translator()


```

```python
# Translate each school in the DataFrame
transformed_df['English_discipline'] = transformed_df['discipline'].apply(lambda x: translate_discipline(x, translator))
```
